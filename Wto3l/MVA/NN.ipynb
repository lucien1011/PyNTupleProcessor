```python
#Import files and select variables

input_files = ['/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WpTo3l_ZpM15_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WpTo3l_ZpM20_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WpTo3l_ZpM30_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WpTo3l_ZpM45_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WpTo3l_ZpM60_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WmTo3l_ZpM15_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WmTo3l_ZpM20_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WmTo3l_ZpM30_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WmTo3l_ZpM45_PS.root',
               '/cmsuf/data/store/user/t2/users/nikmenendez/skimmed/signal/MVA/WmTo3l_ZpM60_PS.root']
treename = 'passedEvents'

vars_in = ['Lep1_pt','Lep1_eta','Lep1_phi','Lep1_iso','Lep1_id',
           'Lep2_pt','Lep2_eta','Lep2_phi','Lep2_iso','Lep2_id',
           'Lep3_pt','Lep3_eta','Lep3_phi','Lep3_iso','Lep3_id',
           'met','met_phi','dR12','dR13','dR23','c0','c1','c2','c3']

vars_test = ['Lep1_pt','Lep1_eta','Lep1_phi','Lep1_iso','Lep1_id',
           'Lep2_pt','Lep2_eta','Lep2_phi','Lep2_iso','Lep2_id',
           'Lep3_pt','Lep3_eta','Lep3_phi','Lep3_iso','Lep3_id',
           'met','met_phi','dR12','dR13','dR23']

vars_check = ['c0','c1','c2','c3']

#vars_in = list(set(vars_in))

print "... Opening input files:"

arrs_sig = [None]*(len(input_files))

print "Opening file: ",input_files[0]
arrs_sig[0] = uproot.open(input_files[0])[treename]
signal = arrs_sig[0].pandas.df(vars_in)

for i in range(1,len(input_files)):
    print "Opening file: ", input_files[i]
    arrs_sig[i] = uproot.open(input_files[i])[treename]
    temp_sig = arrs_sig[i].pandas.df(vars_in)
    signal = signal.append(temp_sig)
    
#print signal[vars_in]
```

```python
# Normalize Features

def min_max_scaling(df):
    # copy the dataframe
    df_norm = df.copy()
    # apply min-max scaling
    for column in df_norm.columns:
        print column + ": min = %f, max = %f" % (df_norm[column].min(),df_norm[column].max())
        df_norm[column] = (df_norm[column] - df_norm[column].min()) / (df_norm[column].max() - df_norm[column].min())
        
        
    return df_norm
    
# call the min_max_scaling function
signal = min_max_scaling(signal)

#print signal
```

```python
#Shuffle DF and split into training and testing

signal = shuffle(signal)
signal = signal.reset_index(drop=True)
signal_train, signal_test = train_test_split(signal, test_size=0.33, random_state=123456)

# early stopping callback
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10)

# model checkpoint callback
# this saves our model architecture + parameters into dense_model.h5
from keras.callbacks import ModelCheckpoint
model_checkpoint = ModelCheckpoint('dense_model.h5', monitor='val_loss', 
                                  verbose=0, save_best_only=True, 
                                  save_weights_only=False, mode='auto', 
                                  period=1)

#print signal_train[vars_check]
```

```python
#Create model

inputs = Input(shape=(len(vars_test),), name = 'input')
#x = Dropout(.1, name = 'input_dropout_.2')(inputs)
x = Dense(32, name = 'hidden1')(inputs)
x = LeakyReLU(.0, name = 'LeakyReLU1_.0')(x)
x = Dense(64, name = 'hidden2')(x)
x = LeakyReLU(.0, name = 'LeakyReLU2_.0')(x)
#x = Dropout(.2, name = 'hidden2_dropout_.4')(x)
outputs = Dense(4, name = 'output', activation='softmax')(x)

model = Model(inputs=inputs, outputs=outputs)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])

# model = Sequential()
# model.add(Dense(64, input_dim=len(vars_test), activation='relu'))
# model.add(Dense(128, activation='relu'))
# model.add(Dense(4, activation='softmax'))
# # Compile model
# model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(signal_train[vars_test],
                    signal_train[vars_check],
                    epochs=100,
                    batch_size=256,
                    verbose=1,
                    callbacks=[early_stopping], 
                    validation_split=0.25)
```

```python
results = model.evaluate(signal_test[vars_test], signal_test[vars_check])
print results
```
